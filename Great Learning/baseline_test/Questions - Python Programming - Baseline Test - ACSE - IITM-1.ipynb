## Question 1 (10 Marks)

Write a function to remove duplicate elements from the given list. 

`A =  [2, 5, 7, 3, 33, 3, 5, 7]`

***
#### Sample Input/Output:
---
- Input: `A =  [2, 5, 7, 3, 33, 3, 5, 7]`
- Output: `Updated list after removing duplicate elements: [2, 5, 7, 3, 33]`
---

```python
A = [2, 5, 7, 3, 33, 3, 5, 7]

def remove(A): 
    temp_list = [] 
    #### Add your code here ####
    duplicates = []
    for i in A:
        if(i in temp_list):
            duplicates.append(i)
        else:
            temp_list.append(i)

    return temp_list
      
print("Updated list after removing duplicate elements: ", remove(A))
```

## Question 2 (15 Marks)

Write a program that takes input from the user to enter a string (consisting of any characters). Then create and print a dictionary from that string whose keys are the characters of the string and whose values are number of times that character is repeating.


***
#### Sample Input/Output:
---
**Sample 1**

- Input: `Enter a string: AAABBCCCCC##**`
- Output: `Characters with its count:  {'*': 2, 'C': 5, '#': 2, 'B': 2, 'A': 3}`

---
**Sample 2**

- Input: `Enter a string: ''`
- Output: `Characters with its count: {}`
***

```python
s = input('Enter a string: ')

char_count = dict()
#### Add your code here ####
for i in s:
    if i in char_count:
        char_count[i] += 1
    else:
        char_count[i] = 1

print('Characters with its count: ', char_count)
```
